```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_1(data, title):

  
    plt.figure(figsize=(10, 8)) 
    plt.imshow(data, cmap='crest', origin='lower', vmin=0, vmax=100)
    plt.grid(False)
    plt.colorbar(label=title)
    plt.title(f'{title} Map')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.show()
```

```python
def plot_2(data, title):

  
    plt.figure(figsize=(10, 8)) 
    plt.imshow(data, cmap='crest', origin='lower',vmin=0, vmax=1)
    plt.grid(False)
    plt.colorbar(label=title)
    plt.title(f'{title} Map')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.show()
```
